# Multi Head Attention

```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras import layers, Model, utils
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
path = '/Users/akshay/Dropbox/Personal/Data Science/Codes/Projects Non Git/yelp_data/Data/'

files = [f for r,d,f in os.walk(path)][0]
filepaths = [path+i for i in files]
filepaths
```

> The Yelp reviews polarity dataset is constructed by considering stars 1 and 2 negative, and 3 and 4 positive. For each polarity 280,000 training samples and 19,000 testing samples are take randomly. In total there are 560,000 trainig samples and 38,000 testing samples. Negative polarity is class 1, and positive class 2.

```python
df = pd.read_csv(filepaths[2], header=None, encoding='utf8')
df.columns = ['sentiment', 'reviews']
df.head(3)
```

```python
def preprocessing(s):
    s = df.reviews.str.replace(r'\\n',' ', regex=True).str.lower()
    s = s.str.replace(r'.',' ', regex=True)
    s = s.str.replace("[^A-Za-z0-9 ]", "")
    s = s.str.split()
    s = s.apply(lambda x: ' '.join([i.strip() for i in x if len(i)>0]))
    return s
```

```python
df['reviews_clean'] = preprocessing(df.reviews)
df.head()
```

```python
df.reviews_clean.str.split().str.len().hist()

maxlen = 300
plt.axvline(x=maxlen, color='r', linestyle='--')
```

```python
#Tokenizing and padding
tok = Tokenizer(num_words=5000)
tok.fit_on_texts(df.reviews_clean)
seq = tok.texts_to_sequences(df.reviews_clean)
seq_padded = pad_sequences(seq, maxlen=maxlen)
```

```python
#Test train split
X, y = seq_padded, df.sentiment.to_numpy()-1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
[i.shape for i in [X_train, X_test, y_train, y_test]]
```

```python
#Vanilla LSTM Model architecture
inp = layers.Input((300,))
x = layers.Embedding(5000, 50)(inp)
x = layers.LSTM(32, return_sequences=True)(x)
x = layers.LSTM(32)(x)
out = layers.Dense(2, activation='softmax')(x)

model = Model(inp, out)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=32)
```

```python
plt.plot(model.history.history.get('acc'),'g')
```

```python
plt.plot(model.history.history.get('loss'),'r')
```

```python
accuracy_score(y_test,np.argmax(model.predict(X_test), axis=1))
```

```python
#Testing how the model works
sent1 = 'this is a review of Surbhis book. i like it its good and engaging.\
         the book is really really fun'

sent2 = 'this is a review of Surbhis tea. i am not a big fan of her tea. she doesnt put\
         any sugar and so its absolutely tasteless and boring.'

sent3 = 'lets talk about surbhis makeup skills ok? sometimes its over the top and shitty'

samples = pad_sequences(tok.texts_to_sequences([sent1, sent2, sent3]), maxlen=300)
np.where(np.argmax(model.predict(samples), axis=1)==1, 'positive', 'negative') 
```

### Attention based architecture

```python
#Attention layer based architecture
inp = layers.Input((300,))
x = layers.Embedding(5000, 50)(inp)
x = layers.MultiHeadAttention(4, 2, attention_axes=(1,))(x,x)
x = layers.GlobalAveragePooling1D()(x)
out = layers.Dense(2, activation='softmax')(x)

model = Model(inp, out)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=64)
```

```python
plt.plot(model.history.history.get('acc'),'g')
```

```python
plt.plot(model.history.history.get('loss'),'r')
```

```python
accuracy_score(y_test,np.argmax(model.predict(X_test), axis=1))
```

```python
#Testing how the model works
sent1 = 'this is a review of Surbhis book. i like it its good and engaging.\
         the book is really really fun'

sent2 = 'this is a review of Surbhis tea. i am not a big fan of her tea. she doesnt put\
         any sugar and so its absolutely tasteless and boring.'

sent3 = 'lets talk about surbhis makeup skills ok? sometimes its over the top and shitty'

samples = pad_sequences(tok.texts_to_sequences([sent1, sent2, sent3]), maxlen=300)
np.where(np.argmax(model.predict(samples), axis=1)==1, 'positive', 'negative') 
```

### Plotting attention weights as heatmap

```python
#Plotting attention heatmap for a sentence
#sent = 'this restaurent has the best beef in town. it is really really tasty'

idx = 610
sent = df.reviews_clean[idx]
sample = pad_sequences(tok.texts_to_sequences([sent]), maxlen=300)

##Good ones - 200, 420, 500, 570, 610, 650, 700, 1300!, 1400, 1777!!, 1918!, 2704, 2766, 900!!
```

```python
#Fetch layers from the trained model and forward pass to get weights
l = model.layers[:3]
emb = l[1](l[0](sample))
_,w = l[2](emb, emb, return_attention_scores=True)
w2 = np.average(w.numpy(), axis=(1,2)).reshape(1,300)
w2.shape
```

```python
#Plot weights
print('SENTENCE:',df.reviews[idx])
print('')
print('SENTIMENT:')
print(pd.DataFrame(model.predict(sample), columns=['Negative', 'Positive']))

words = np.sum(sample!=0)
tokens = [tok.index_word.get(i) for i in sample[0][-words:]]
attention_df = pd.DataFrame(w2[-words:, -words:], columns=tokens)

plt.figure(figsize=(15,1))
sns.heatmap(attention_df)#, cmap='YlGnBu')
```
