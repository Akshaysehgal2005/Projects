```python
def create_cnn_lstm_model_f_api(input_shape):

    model = Sequential()
    cnn = tensorflow.keras.applications.DenseNet169(include_top=True, 
                                                    weights=None, 
                                                    input_tensor=None, 
                                                    input_shape=input_shape[1:], #<--------
                                                    pooling=None, classes=2)
    
    # define LSTM model
    
    model.add(Input(input_shape, ragged=True))
    
    model.add(tensorflow.keras.layers.TimeDistributed(cnn))
    
    model.add(LSTM(units = 512, 
                   dropout=0.5, 
                   recurrent_dropout=0.3, 
                   return_sequences = True))
    
    model.add(LSTM(units = 512, 
                   dropout=0.5, 
                   recurrent_dropout=0.3, 
                   return_sequences = False))
    
    model.add(Dense(units=NUM_CLASSES, 
                    activation='sigmoid'))#Compile

    model.compile(loss=tensorflow.keras.losses.binary_crossentropy, optimizer='adam', metrics=['accuracy'])
    print(model.summary())

    return model

input_shape = (None, 128, 216, 1)
NUM_CLASSES = 2

model = create_cnn_lstm_model_f_api(input_shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model, utils, applications
```

```python
def cnn_model(image):
    
    inp = layers.Input(image)
    x = layers.Conv2D(8, 3)(inp)
    x = layers.MaxPool2D(3)(x)
    x = layers.Conv2D(16, 3)(x)
    x = layers.MaxPool2D(3)(x)
    x = layers.Conv2D(32, 3)(x)
    x = layers.MaxPool2D(3)(x)
    x = layers.Flatten()(x)
    out = layers.Dense(2, activation='sigmoid')(x)
    
    model = Model(inp, out)
    model.compile(loss='binary_crossentropy', 
                  optimizer='adam', 
                  metrics='accuracy')
    
    return model
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model, utils, activations, optimizers, backend
```

```python
X = np.random.random((10,100,300,1))
y = np.random.random((10,3))
```

```python
filter_size = (3,3)
filters = 32
pool = 2

input_layer = layers.Input(shape=(100,300,1))

conv_extractor = layers.Conv2D(filters,filter_size, activation='relu')(input_layer)
conv_extractor = layers.MaxPooling2D(pool_size=(pool, pool))(conv_extractor)
conv_extractor = layers.Conv2D(filters,filter_size, activation='relu')(conv_extractor)
conv_extractor = layers.MaxPooling2D(pool_size=(pool, pool))(conv_extractor)

#conv_extractor = layers.Reshape(target_shape=(100 // (pool ** 2), (100 // (pool ** 2)) * filters))(conv_extractor)
shape = ((100 // 4), (300 // 4) * 32)
#conv_extractor = layers.Dense(512, activation='relu')(conv_extractor)
conv_extractor = layers.Reshape(target_shape=(23,2336))(conv_extractor)

gru_1 = layers.GRU(512, return_sequences=True)(conv_extractor)
gru_1b = layers.GRU(512, return_sequences=True, go_backwards=True)(conv_extractor)
gru1_merged = layers.add([gru_1, gru_1b])
gru_2 = layers.GRU(512, return_sequences=True)(gru1_merged)
gru_2b = layers.GRU(512, return_sequences=True, go_backwards=True)(gru1_merged)

x = layers.concatenate([gru_2, gru_2b])   # move concatenate layer aside
x = layers.Flatten()(x)
inner = layers.Dense(30, activation=layers.LeakyReLU())(x)
inner = layers.Dense(10, activation=layers.LeakyReLU())(inner)
inner = layers.Dense(3, activation=layers.LeakyReLU())(inner)
inner = layers.Lambda(lambda x: backend.round(x))(inner)


model = Model(input_layer,inner)
model.compile(loss = "MeanSquaredError", optimizer = optimizers.Adam(2e-4), metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X, y, epochs = 5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
%matplotlib notebook

#create image with format (time,x,y)
data = np.random.random((100,))#10,10)

#setup figure
fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(1,1,1)


# #set up viewing window (in this case the 25 most recent values)
repeat_length = (np.shape(data)[0]+1)//4

ax.set_xlim([0,repeat_length])
ax.set_ylim([-2,2])

# #set up list of images for animation

im, = ax.plot([], [])

def func(n):
    im.set_xdata(np.arange(n))
    im.set_ydata(data[0:n])
    if n>repeat_length:
        lim = ax.set_xlim(n-repeat_length, n)
    else:
        lim = ax.set_xlim(0,repeat_length)
    return im

ani = animation.FuncAnimation(fig, func, frames=data.shape[0], interval=30, blit=False)

plt.show()

ani.save('animation.gif',writer='pillow', fps=20)
```
